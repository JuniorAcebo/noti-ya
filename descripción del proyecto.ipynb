# Proyecto: Aplicación de Noticias Sudamericanas

## Descripción del proyecto

Este proyecto consiste en el desarrollo de una aplicación móvil (APK) de noticias enfocada en la región sudamericana, incluyendo países como Bolivia, Brasil, Chile, Perú, Colombia, etc. La aplicación permite visualizar noticias actualizadas en tiempo real gracias al consumo de una API externa, lo que significa que la información no está guardada en el celular, sino que se obtiene directamente desde internet cada vez que el usuario accede a la app.

Para lograr esta funcionalidad, se utilizó la librería Retrofit, la cual permite conectar la aplicación con servicios web (APIs) de forma sencilla y organizada. Además, se empleó OkHttp para gestionar las conexiones a internet (peticiones HTTP) y Gson para convertir los datos recibidos en formato JSON a objetos que la aplicación puede entender y mostrar correctamente.

El desarrollo se realizó en el entorno de Android Studio utilizando el lenguaje Java, configurado con el JDK correspondiente. La aplicación fue diseñada para ofrecer una experiencia simple e intuitiva al usuario, permitiendo consultar noticias de diferentes países de manera rápida y eficiente.

Este proyecto fue desarrollado en el año 2025, durante el mes de marzo, con una duración aproximada de 1 mes, demostrando habilidades en el consumo de APIs, desarrollo móvil y manejo de datos en tiempo real.
